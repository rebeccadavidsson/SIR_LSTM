## (Experimental): Relationship of OxCGRT index and parameter values
Government responses will impact on parameter values of ODE models about 14 days later. Here, we will investigate on the relationship of [stringency index (from OxCGRT)](https://github.com/OxCGRT/covid-policy-tracker/blob/master/documentation/index_methodology.md) and reproduction number/model parameters.


```python
# SIRF model derived ODE parameters for 14 countries
# df = pickle.load(open('figures/pickles/df_14_countries.p', 'rb'))

df = analyser.track()
param_cols = ["Rt", *cs.SIRF.PARAMETERS]
oxcgrt_cols = oxcgrt_data.OXCGRT_VARS[:]
cols = param_cols + oxcgrt_cols
delay = 14

for col in param_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df[col] = df.groupby("Country")[col].shift(0 - delay)

df = df.dropna(how="any")
df = df.loc[:, ["Date", "Country", *cols]]
track_df = df.copy()
```

```python
countries = df["Country"].unique()
print(countries)
```

```python
def crosscorr(datax, datay, lag=0):
    """
    Calcalates the correlation between x and y, where
    y is a shifted column, shifted by n number of days (= lag).
    Returns the correlation between column x and y
    """
    return datax.corr(datay.shift(lag))
```

```python
def change_point_detection(corr_list, country, model="l2", default_days=14, breaks=6):
    """
        Calculate the windows of 
        Models: "l2",  "l1", "rbf", "linear", "normal", "ar"
    """
    
    corr = pd.DataFrame(corr_list)
    algo = rpt.Binseg(model=model).fit(corr)
    bkps = algo.predict(n_bkps=breaks)
    
    window = corr_list[bkps[0]:bkps[1]]
    
    # Get location of LOWEST correlation value (= strongest negative correlation)
    delay_days = np.argmin(window) + bkps[0]
    color = "blue"
    if bkps[0] >= 30:
        print("Delay not found, delay set to 14")
        delay_days = default_days
        color = "red"
        
    fig, ax = rpt.show.display(corr, bkps);
    ax[0].set_title("Change point analysis of Rt in " + country + "," + str(delay_days), fontsize=14)
    ax[0].set_ylabel("Rt")
    ax[0].set_xlabel("Days")
    ax[0].axvline(delay_days, color=color)
    
    return delay_days
```

## Set the window to search for delay effect
Minimum number of days is set to zero. 

```python
# Max number of days to delay effect
max_delay_days = 23

param = "Rt"
```

```python
corr_countries = {}

for country in countries:
    df_country = df[df["Country"] == country] 
    
    # Calculate mean of the given parameter for each day
    grouped = df_country.groupby("Date").mean()
    
    # Calculate increase or decrease in Rt for each day
    grouped = grouped.reset_index()
    grouped["Rt_change"] = grouped[param].diff(1)
    start_date = find_start_date(grouped["Rt_change"].values)
    
    # Select subset of data only where Rt started to increase
    df_country = df_country.query("Date>" +  start_date)

    # Calculate correlation for several days of delay until max_delay is reached
    corr_list = [crosscorr(grouped[param], grouped["Stringency_index"], lag=i) for i in range(0, max_delay_days)]
    
    # Append to dict for later use
    corr_countries[country] = corr_list
    
```

```python
# Calculate increase or decrease in Rt for each day
grouped = df_country.groupby("Date").mean()
grouped = grouped.reset_index()
grouped[param + "_change"] = grouped[param].diff(1)
grouped.plot(x="Date", y=param + "_change", title="Change in chosen parameter over time in " + country);
```

```python
total_delay_days = []
for country in corr_countries:
    
    # Set default number of days for countries where delay cannot be calculated
    default_delay = 14
    
    delay_days = change_point_detection(corr_countries[country], country, model="l2", default_days=14, breaks=2)
    total_delay_days.append(delay_days)
    
```

### In this way, days of delay is not always calculated correcly;
It is expected to see a change in correlation coefficient between 10 and 20 days. However, this approach is not applicable to all countries. An example is Sweden, where the change in Rt is very minimal. 
A solution would be to look at the change in gradients, to possibly detect a rapid change in gradients:


```python
gradient = np.gradient(corr_countries[country])
gradients[country] = gradient
plt.plot(gradient, label=country)
plt.legend(loc="upper left")
plt.title("Gradient of correlation coefficient for one country (example)")
plt.xlabel("Days of delay")
plt.ylabel("Correlation coefficint")
plt.show()


gradients = {}
for country in corr_countries:
    gradient = np.gradient(corr_countries[country])
    gradients[country] = gradient
    plt.plot(gradient, label=country)
plt.title("Gradient of correlation coefficient for all countries")
plt.xlabel("Days of delay")
plt.ylabel("Correlation coefficint")
plt.legend()
plt.show()

```

Again, changes are very, very minimal.

```python
from itertools import cycle
min_delay_days = 10
max_delay_days = 21

total_days = []
corr_df = pd.DataFrame.from_dict(corr_countries, orient="columns")
nrow, ncol = 7, 2
colors = map('C{}'.format, cycle(range(nrow*ncol)))
fig, axs = plt.subplots(nrow,ncol,figsize=(ncol*5.5, nrow*4), constrained_layout=True);

for country, ax in zip(countries, axs.flatten()):
    corr_df["Diff_" + country] = corr_df[country].diff(1)
    
    # Maximum difference
    days = np.argmax(abs(corr_df["Diff_" + country].values[min_delay_days:max_delay_days]))
    
    # Check for outliers
    if days + min_delay_days >= max_delay_days - 1:
        days = 4
    total_days.append(days + min_delay_days)
    ax.bar(range(0, len(corr_df[min_delay_days:max_delay_days])), abs(corr_df["Diff_" + country][min_delay_days:max_delay_days]), label=country)
    ax.set_title(country)
#     ax.axvline(days)    
plt.show()
```
